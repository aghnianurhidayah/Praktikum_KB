##### 3. Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka

```python
jml_record = data.size
jml_attribute = len(data.columns)
nama_attribute = data.columns.tolist()
jml_attribute_angka = len(data.select_dtypes('int').columns)

info = {
    "Jumlah record" : jml_record,
    "Jumlah attribute" : jml_attribute,
    "Nama attribute" : nama_attribute,
    "Jumlah attribute bertipe angka" : jml_attribute_angka
}

for key, value in info.items():
    print(key, ":", value)
```
