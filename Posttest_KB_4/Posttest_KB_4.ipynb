### Data Cleaning Nilai Null

Untuk memeriksa jumlah nilai null yang ada pada dataset gunakan perintah berikut

```python
data_manipulation.isna().sum()
```

Pada record yang null di kolom 'age' akan diisi dengan rata-rata dari jumlah record pada attibute 'age'

```python
data_manipulation[['age']] = data_manipulation[['age']].fillna(data_manipulation[['age']].mean())
```

Sehingga nilai null yang ada pada kolom 'age' tidak ada

```python
data_manipulation.isna().sum()
```

### Data Cleaning Nilai Duplikat

Untuk memeriksa jumlah nilai duplikat yang ada pada dataset gunakan perintah berikut

```python
data_manipulation.duplicated().sum()
```

Pada data_manipulation ditemukan bahwa terdapat 6 baris/record yang memiliki bernilai sama/duplikat. Sehingga untuk menghilangkan nilai duplikat, perlu untuk mengahapus nilai yang sama. 

```python
data_manipulation.drop_duplicates(inplace=True, ignore_index=True)
```

Setelah nilai yang sama dihapus, jumlah nilai duplikat yang ada pada dataset adalah 0

```python
data_manipulation.duplicated().sum()
```

### Mengubah Tipe data attribute angka

Karena attribute 'age' memiliki tipe data float, maka kita dapat mengubah tipe data tersebut menjadi integer

```python
data_manipulation.dtypes
```

Untuk mengubah tipe data, gunakan fungsi 'astype()'

### Data Transform

Sebelum data ditransform, cek tipe data dari tiap attribute. Data yang memiliki tipe data numerik dapat digunakan untuk transform data

```python
data_manipulation.dtypes
```

Attribute yang dapat digunakan adalah 'age', 'balance', dan 'estimated_salary'

### Standarisasi

```python
ss = StandardScaler()
data_standard = pd.DataFrame(data_manipulation[['age','balance','estimated_salary']])
data_standard[['age','balance','estimated_salary']] = ss.fit_transform(data_standard[['age', 'balance', 'estimated_salary']])
```

Setelah ditransform menggunakan teknik strandarisasi, nilai pada attribute 'age', 'balance', dan 'estimated_salary' memiliki skala 0 hingga 1

```python
data_standard
```

### Normalisasi

```python
mm = MinMaxScaler()
data_normal = pd.DataFrame(data_manipulation[['age','balance','estimated_salary']])
data_normal[['age','balance','estimated_salary']] = mm.fit_transform(data_normal[['age','balance','estimated_salary']])
```

```python
data_normal
```

Setelah ditransform menggunakan teknik normalisasi, nilai min pada attribute 'age', 'balance', dan 'estimated_salary' adalah 0 dan nilai max nya adalah 1

```python
data_normal.describe()
```

### One Hot Encoding

Pada attribute 'gender', kita dapat menggunakan fungsi one hot encoder untuk mengubah nilai kategorikal menjadi nilai boolean (True/False)

```python
ohe = OneHotEncoder(sparse_output=False)
oh_encoded = ohe.fit_transform(data_manipulation[['gender']])
```

```python
gender_encoded = pd.DataFrame(oh_encoded,
                      columns=['female', 'male'])
```

Setelah berhasil diubah, maka nilai pada attribute 'gender' terpecah menjadi 'male' dan 'female' dan berisi nilai boolean yaitu 0 dan 1. 

```python
gender_encoded
```

Kemudian gabungkan data 'gender_encoded' ke dalam data_manipulation dan hapus kolom 'gender'

```python
data_manipulation = pd.concat([data_manipulation, gender_encoded], axis=1)
```

```python
data_manipulation.drop(columns=['gender'], inplace=True)
```

Sehingga, pada data_manipulation, kolom gender diganti menjadi female dan male yang bernilai boolean

```python
data_manipulation
```

### Split Data

Pada dataset ini, attribute 'churn' merupakan label/target, sehingga variabel y akan berisi 'churn' dan variabel x akan berisi semua attribute selain 'churn'

```python
x = data_manipulation.drop(columns='churn')
y = data_manipulation['churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```
