```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np

data = pd.read_csv('Bank Customer Churn Prediction.csv')
```

```python
data
```

### Pie Chart

```python
country = data.country.value_counts()

country.plot.pie(autopct='%.2f%%')

plt.title('Jumlah Nasabah')
plt.show()
```

Pie Chart di atas menunjukkan nasabah yang berasal dari negara Perancis adalah sebagian dari total nasabah yaitu sebanyak 50.14%, sedangkan nasabah dari negara jerman sebanyak 25.09% dan nasabah dari negara spayol sebanyak 24.77%

### Bar Plot

```python
salary = data.groupby('country').mean()

barplot = sns.barplot(x=salary.index,
                    y=salary['estimated_salary'])

barplot.set(xlabel='Country',
           ylabel='Estimated Salary',
           title='Estimated Salary by Country')

plt.show()
```

Walaupun nasabah didominasi oleh penduduk Perancis tetapi rata-rata gaji tertinggi dimiliki oleh nasabah yang berasal dari Jerman, kemudian disusul Perancis, dan Spanyol

### Histogram

```python
plt.hist(data['balance'], range=(0, 250000))
plt.title('Saldo Nasabah Bank')
plt.ylabel('Frekuensi')
plt.xlabel('Jumlah Saldo')
plt.show()
```

Histogram di atas menunjukkan bahwa lebih dari 3500 nasabah tidak memiliki saldo atau saldonya bernilai 0. Sekitar 2000 lebih nasabah memiliki saldo 100.000 sampai 125.000. 1250 nasabah memiliki saldo di bawah 100.000. Dan 1250 nasabah lainnya memiliki saldo di atas 125.000.  

### Heatmap

```python
cor_matrix = data.loc[:, 'tenure':'estimated_salary'].corr()
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.show()
```

Visualisasi di atas menunjukkan bahwa antara variabel tenure, balance, products_number, credit_card, active_member dan estimated_salary mempunyai hubungan yang lemah karena korelasinya menunjukkan jauh dari 1 bahkan ada beberapa yang korelasinya minus.
